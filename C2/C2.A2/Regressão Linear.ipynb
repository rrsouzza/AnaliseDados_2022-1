```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

# Ex. 1
# a)

```python
df_massaMuscular = pd.read_csv('DataSets/massa_muscular_x_idade.csv', sep=',')

#df_massaMuscular.plot.scatter(x='Idade', y='Massa_muscular')
#plt.tight_layout()
#plt.show()

##################################################################

#predictors = ['Idade']
#outcome = 'Massa_muscular'
#model = linear_model.LinearRegression()
#model.fit(df_massaMuscular[predictors], df_massaMuscular[outcome])
#print(f'Intercept: {model.intercept_:.3f}')
#print(f'Coefficient Exposure: {model.coef_[0]:.3f}')

##################################################################

predictors = ['Idade']
outcome = 'Massa_muscular'
model = linear_model.LinearRegression()
model.fit(df_massaMuscular[predictors], df_massaMuscular[outcome])

fitted = model.predict(df_massaMuscular[predictors])
residuals = df_massaMuscular[outcome] - fitted
ax = df_massaMuscular.plot.scatter(x='Idade', y='Massa_muscular', figsize=(4, 4))
ax.plot(df_massaMuscular.Idade, fitted)

for x, yactual, yfitted in zip(df_massaMuscular.Idade, df_massaMuscular.Massa_muscular, fitted):
    ax.plot((x, x), (yactual, yfitted), '--', color='C1')
plt.tight_layout()
plt.show()
```

```python
df_rendaFamiliar = pd.read_csv('DataSets/renda_familia_alimentacao.csv', sep=',')
predictors2 = ['Renda_Familiar_X']
outcome2 = 'Gasto_Alimentacao_Y'
model2 = linear_model.LinearRegression()
model2.fit(df_rendaFamiliar[predictors2], df_rendaFamiliar[outcome2])

fitted2 = model2.predict(df_rendaFamiliar[predictors2])
residuals2 = df_rendaFamiliar[outcome2] - fitted2
ax = df_rendaFamiliar.plot.scatter(x='Renda_Familiar_X', y='Gasto_Alimentacao_Y', figsize=(4, 4))
ax.plot(df_rendaFamiliar.Renda_Familiar_X, fitted2)

for x, yactual, yfitted2 in zip(df_rendaFamiliar.Renda_Familiar_X, df_rendaFamiliar.Gasto_Alimentacao_Y, fitted2):
    ax.plot((x, x), (yactual, yfitted2), '--', color='C1')
plt.tight_layout()
plt.show()
```
