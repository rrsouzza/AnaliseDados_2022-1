```python
#Ex.5
def minutosBuilder(faixas, frequencia):
    finalArray = []
    
    index = 0
    while (index < len(faixas)):
        x = 0
        while (x < frequencia[index]):
            finalArray.append(faixas[index] + 50)
            # Para cálculo de média de minutos assistidos, é considerado os valores médios de minutos ('se está 
            # calculando a faixa de 300-400, é considerado 350 minutos. Se 500-600, 550 minutos, etc.') já que
            # não se tem os valores reais de minutos assistidos por cada aluno do ensino médio.
            x += 1
        index += 1
    return finalArray

##################################################################

df_ensinoMedio = pd.DataFrame({
    'Faixas': [
        300, 400, 500, 600, 700, 800, 900, 1000, 1100
    ],
    'Frequencia': [
        14, 46, 58, 76, 68, 62, 48, 22, 6
    ]    
})
df_ensinoMedio_minutos = pd.DataFrame({
    'Minutos': minutosBuilder(df_ensinoMedio['Faixas'], df_ensinoMedio['Frequencia'])
})


df_universitarios = pd.DataFrame({
    'Minutos': [
        350, 560, 580, 710, 945, 880, 760, 640, 660, 820, 
        775, 910, 920, 850, 810, 790, 890, 685, 730, 850, 
        745, 640, 1010, 420, 770, 850, 915, 840, 930, 895
    ]
})

##################################################################

print('Estatísticas da quantidade de minutos assistidos pelos alunos do ensino médio:\n\n', df_ensinoMedio_minutos.describe(), '\n\n')
print('Estatísticas da quantidade de minutos assistidos pelos alunos universitários:\n\n', df_universitarios.describe(), '\n\n')

##################################################################

labels = ['300-400', '400-500', '500-600', '600-700', '700-800', '800-900', '900-1000', '1000-1100', '1100-1200']

plt.rcParams['font.size'] = '16'
plt.rcParams["figure.figsize"] = (18, 6)
plt.title('Histograma Alunos Ensino Médio')

plt.bar(labels, df_ensinoMedio['Frequencia'], width=0.7, color='brown', edgecolor='black')
plt.xlabel('Minutos Assistidos', fontsize=20)
plt.ylabel('Frequência de Alunos', fontsize=20)
plt.show()

##################################################################

freq_universitarios = pd.cut(df_universitarios['Minutos'], 10)
print('Tabela de frequência dentre os universitários:\n\n', freq_universitarios.value_counts().sort_index(), '\n\n')

ax = df_universitarios['Minutos'].plot.hist(figsize=(18, 6), width=30)

ax.set_xlabel('Minutos Assistidos')
plt.tight_layout()
plt.show()

##################################################################

print('Ao analisar os dados, podemos concluir: \n')
print('-Os alunos universitários assistem televisão, em média, mais tempo que os alunos do ensino médio.\n')
print('-O desvio padrão calculado indica que os alunos do ensino médio assistem televisão durante uma quantidade de minutos mais variada (menos ou mais que a média) que os alunos universitários.\n')
print('-A quantidade de alunos que assistem televisão por MAIS tempo (baseado na média) é maior que a quantidade de alunos que assitem por menos tempo, tanto entre os alunos do ensino médio quanto entre os alunos universitários.\n')
```

```python
#Ex. 6
def quilometrosBuilder(faixas, frequencia):
    finalArray = []
    
    index = 0
    while (index < len(faixas)):
        x = 0
        while (x < frequencia[index]):
            if faixas[index] == 0 or faixas[index] == 2:
                finalArray.append(faixas[index] + 1)
            elif faixas[index] == 4 or faixas[index] == 8 or faixas[index] == 12:
                finalArray.append(faixas[index] + 2)
            elif faixas[index] == 16:
                finalArray.append(faixas[index] + 3)
                # Para cálculo de média de quilômetros corridos, é considerado os valores médios de quilômetros
                # ('se está calculando a faixa de 0-2, é considerado 1 quilômetro. Se 4-8, 6 quilômetros, etc.')
                # já que não se tem os valores reais de quilômetros corridos por cada aluno.
            x += 1
        index += 1
    return finalArray

##################################################################

# a)

df_frequencia = pd.DataFrame({
    'Faixas': [
        0, 2, 4, 8, 12, 16
    ],
    'Frequencia Anterior': [
        442, 200, 130, 34, 10, 3
    ],
    'Frequencia Posterior': [
        80, 200, 280, 179, 43, 8
    ]
})

labels = ['0-2', '2-4', '4-8', '8-12', '12-16', '16-22']

##################################################################

df_quilometros_anterior = pd.DataFrame({
    'Quilometros': quilometrosBuilder(df_frequencia['Faixas'], df_frequencia['Frequencia Anterior'])
})

df_quilometros_posterior = pd.DataFrame({
    'Quilometros': quilometrosBuilder(df_frequencia['Faixas'], df_frequencia['Frequencia Posterior'])
})

##################################################################
#Histograma de Antes

plt.rcParams['font.size'] = '16'
plt.rcParams["figure.figsize"] = (18, 6)
plt.title('Histograma Frequência Antes do Treinamento')

plt.bar(labels, df_frequencia['Frequencia Anterior'], width=0.7, color='brown', edgecolor='black')
plt.xlabel('Quilômetros Atingidos', fontsize=20)
plt.ylabel('Frequência de Alunos', fontsize=20)
plt.show()

##################################################################
#Histograma de Depois

plt.rcParams['font.size'] = '16'
plt.rcParams["figure.figsize"] = (18, 6)
plt.title('Histograma Frequência Depois do Treinamento')

plt.bar(labels, df_frequencia['Frequencia Posterior'], width=0.7, color='green', edgecolor='black')
plt.xlabel('Quilômetros Atingidos', fontsize=20)
plt.ylabel('Frequência de Alunos', fontsize=20)
plt.show()

##################################################################
# b)

print('Estatísticas sobre a frequência dos alunos antes do treinamento:\n', df_quilometros_anterior['Quilometros'].describe(), '\n\n')
print('Estatísticas sobre a frequência dos alunos após o treinamento:\n', df_quilometros_posterior['Quilometros'].describe(), '\n\n')

##################################################################

print('Ao analisar os dados, podemos concluir: \n')
print('-Após o treinamento, menos alunos participaram do teste.\n')
print('-Após o treinamento, a média de quilômetros corridos pelos alunos aumentou. \n Antes do treinamento a média foi de 2.88km por aluno, e após o treinamento a média foi de 6.20km.\n')
print('-A quantidade de alunos que correram de 2 a 12 quilômetros aumentou drasticamente após o treinamento.\n')
print('-A quantidade de alunos que correram apenas 0 a 2 quilômetros diminuiu drasticamente após o treinamento.\n')
```

```python
#Ex. 9

df_producao = pd.read_csv('DataSets/estados_producao_industrial.csv', sep=';')

# a)

#Gráfico de Dispersão de Vendas x log(V)
axV = df_producao.plot.scatter(x='logV', y='V', figsize=(4, 4), marker='$\u25EF$')
axV.set_xlabel('log Vendas')
axV.set_ylabel('Vendas')
axV.axhline(0, color='grey', lw=1)
axV.axvline(0, color='grey', lw=1)
plt.tight_layout()
plt.show()

#Gráfico de Dispersão de Pessoas x log(P)
axP = df_producao.plot.scatter(x='logP', y='P', figsize=(4, 4), marker='$\u25EF$')
axP.set_xlabel('log Pessoas')
axP.set_ylabel('Pessoas')
axP.axhline(0, color='grey', lw=1)
axP.axvline(0, color='grey', lw=1)
plt.tight_layout()
plt.show()

#Gráfico de Dispersão de log(V) x log(P)
axVP = df_producao.plot.scatter(x='logV', y='logP', figsize=(4, 4), marker='$\u25EF$')
axVP.set_xlabel('log Vendas')
axVP.set_ylabel('log Pessoas')
axVP.axhline(0, color='grey', lw=1)
axVP.axvline(0, color='grey', lw=1)
plt.tight_layout()
plt.show()

#Gráfico de Dispersão de Vendas x Pessoas
axPV = df_producao.plot.scatter(x='P', y='V', figsize=(4, 4), marker='$\u25EF$')
axPV.set_xlabel('Pessoas')
axPV.set_ylabel('Vendas')
axPV.axhline(0, color='grey', lw=1)
axPV.axvline(0, color='grey', lw=1)
plt.tight_layout()
plt.show()

print('Ao analisar os gráficos, podemos observar que o número de vendas cresceu conforme o número de pessoas/funcionários aumentou.\n')

##################################################################
# b)

print('Coeficiente de Correlação de Pearson:')
print(round(df_producao['logV'].corr(df_producao['logP'], method='pearson'), 3), '\n')
print('Por estar bem próximo de 1, essa é uma correlação considerada alta.\n')
```
